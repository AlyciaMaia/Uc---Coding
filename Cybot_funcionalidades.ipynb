```python
#CadastroEmpresa
print("Cadastro")
empresa = input("Nome da empresa: ")
cnpj = input("Insira o CNPJ: ")
emailEmp = input("Insira o email de contato: ")
senha = input("Crie sua senha: ")
senhaConfirmada = input("Digite sua senha novamente: ")

#verificaçãoSenha
while senhaConfirmada != senha:
    print("As senhas precisam ser iguais. Digite novamente")
    senhaConfirmada = input("Digite sua senha novamente: ")

listaEmp = []
empresa = {"nomeEmpresa:": empresa, "cnpj": cnpj, "emailEmpresa": emailEmp, "senha": senha, "senhaConfirmada": senhaConfirmada}
listaEmp.append(empresa)
print(listaEmp)

```

```python
#Sistema Monitorado (Relatorio)

def gerarRelatorio(sistemaMonitorado):
  mensagem = f"""RELATÓRIO DE SEGURANÇA
  Informações sobre o sistema:
  - Sistema ID: {sistemaMonitorado["idSistema"]}
  - Tipo do Sistema: {sistemaMonitorado["tipoSistema"]}
  - Empresa Responsável: {sistemaMonitorado["idEmpresa"]}
  - Status do Monitoramento: {sistemaMonitorado["statusMonitoramento"]}

  Análise:
  O sistema monitorado encontra-se atualmente em estado '{sistemaMonitorado["statusMonitoramento"]}'.
  Análise recente identificou: {sistemaMonitorado["vulnerabilidades"]}.

  Recomendação:
  Manter o monitoramento contínuo para garantir que nenhuma nova ameaça seja encontrada."""

  print(mensagem)


#Gerando Relatório
sistemaMonitorado = {"idSistema": "SYS001", "tipoSistema": "Servidor", "statusMonitoramento": "Ativo", "idEmpresa": "EMP123",
                     "vulnerabilidades": "Nenhuma vulnerabilidade detectada", "recomendacao": "Recomenda-se manter o monitoramento ativo e realizar análises regulares."}

gerarRelatorio(sistemaMonitorado)





```

```python

#InteraçãoIA

sistemaMonitorado = {"idSistema": "SYS001", "tipoSistema": "Servidor", "statusMonitoramento": "Ativo", "idEmpresa": "EMP123",
                     "vulnerabilidades": "Movimentação não reconhecida detectada no mês de abril, envolvendo acesso externo em horário incomum.",
                     "recomendacao":"""
              Alerta:
              Foi detectada uma movimentação não reconhecida no sistema SYS001 durante o mês de abril,
              caracterizada por um acesso externo em horário incomum. Esse comportamento indica possível tentativa de acesso não autorizado ao servidor.

              Recomendação:
              Revise imediatamente os registros de auditoria (logs) para identificar quais usuários foram afetados.
              Redefina as senhas das contas envolvidas e implemente autenticação multifator (MFA) para fortalecer o acesso.
              Considere configurar alertas automáticos para acessos fora do horário padrão da empresa e restringir IPs não autorizados.""" }

print("Sugestões Rápidas")
print("1 - Ver status do sistema")
print("2 - Ver vulnerabilidades")
print("3 - Ver recomendação da IA")

opcao = input("o que você precisa?")

match opcao:
  case "1":
     print(f"Status do sistema {sistemaMonitorado["idSistema"]}:", sistemaMonitorado["statusMonitoramento"])
  case "2":
    print("Vulnerabilidades:", sistemaMonitorado["vulnerabilidades"])
  case "3":
    print("Sugestão da IA:", sistemaMonitorado["recomendacao"])
  case _:
    print("Opção inválida. Tente novamente.")
```

```python
#LogsDoSistema
lista_logs = [{"data": "12/04/2025 02:15", "evento": "Acesso suspeito",
        "descricao": "Tentativa de acesso externo em horário incomum."},

        {"data": "10/04/2025 14:22", "evento": "Login bem-sucedido",
        "descricao": "Usuário ADMIN acessou o sistema."},

        {"data": "08/04/2025 18:40", "evento": "Alteração do sistema",
        "descricao": "Configurações do firewall foram atualizadas."}]

print("HISTÓRICO DE LOGS DO SISTEMA")
print("Escolha a opção para filtrar o histórico")
print("1 - Ver todos os logs")
print("2 - Ver apenas acessos suspeitos")
print("3 - Ver alterações do sistema")
opcao = input("Opção escolhida: ")

match opcao:
  case "1":
    for log in lista_logs:
      print(f"{log["data"]} - {log["evento"]}: {log["descricao"]}")
  case "2":
      for log in lista_logs:
               if log["evento"] == "Acesso suspeito":
                print(f"{log["data"]} - {log["evento"]}: {log["descricao"]}")
  case "3":
    for log in lista_logs:
                if log["evento"] == "Alteração do sistema":
                  print(f"{log["data"]} - {log["evento"]}: {log["descricao"]}")
  case _:
    print("Opção inválida.")
```
